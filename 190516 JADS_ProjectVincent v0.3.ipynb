```python
import pandas as pd
import re
# import glob
```

```python
# create a dataframe from the relevant lines of the PDF
def create_df(strContent):
    lines = strContent.split('\n')
    matched_lines = []
    for line in lines:
        if re.match("\d{6}", line):
            matched_lines.append(line)
    df = pd.DataFrame(matched_lines)
    return df

pdf = (parsedPDF['content'])
df_creator = create_df(pdf)
df_creator.head()
```
